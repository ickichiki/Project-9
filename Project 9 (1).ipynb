One of my good friends Clara absolutely loves this show called Taskmaster which she got me hooked on my sophomore year. The show is hosted by two comedians, Greg Davis and Alex Horne, and the premise is that they invite 5  comedians onto the show each season and challenge them to do some ridiculous tasks. Some examples of tasks include: not blinking for the longest time, making a Swedish man blush, and getting to a microwave in time to stop it from making soup too hot by taking as few steps as possible.

On Data is Plural, I found data compiled by a Reddit user who went through the website Taskmaster.info and created a number of datasets containing general demographic information about the contestants, as well as more specific information like how many times they cursed over the course of their respective series. 

One thing the show has been criticised for is its lack of representation of female comedians, with female comedians usually making up 1, at most 2, of the five contestants. At times, it appears that Greg Davis, who orchestrates the show, seems to openly favor the male contestants, what with challenges being things like "Who can tie a tie the fastest?" or having a tendency to give more points to male contestants than female contestants. I am curious to see if this is true and if the dataset corroborates that Greg has this tendency against women.

```python
import requests
import json

import pandas as pd
import numpy as np
```

```python
people_url = 'https://prettygr.im/tdlm/taskmaster/people.json?_labels=on'
```

```python
people_response = requests.get(people_url)
```

```python
people_json = people_response.json()
```

```python
people_json
```

```python
taskmaster_people = pd.DataFrame(columns=["id","Series", "Seat", "Name","dob","gender","hand","team","champion","TMI"])
```

```python
for row in people_json['rows']:
    tm_id = row['id']
    series = row['series']
    if series is not None:
        series = series['label']
    seat = row['seat']
    name = row['name']
    dob = row['dob']
    gender = row['gender']
    hand = row['hand']
    team = row['team']
    if team is not None:
        team = team['label']
    champion = row['champion']
    tmi = row['TMI']
    
    new_row = {"id":tm_id,"Series": series, "Seat": seat, "Name":name, "dob":dob, "gender":gender,"hand":hand,"team":team, "champion":champion,"TMI":tmi}
    
    taskmaster_people = pd.concat([taskmaster_people, pd.DataFrame([new_row])], ignore_index=True)
```

```python
taskmaster_people
```

First, I'm going to see how many female contestants there are versus male.

```python
taskmaster_people['gender'].value_counts()
```

Yeesh, not looking good.

What about the winning proportion? How many women have won a series versus a man?

```python
women = taskmaster_people[taskmaster_people['gender']=='F']
men = taskmaster_people[taskmaster_people['gender']=='M']
```

```python
women['champion'].value_counts()
```

```python
men['champion'].value_counts()
```

Okay, of 9 seasons, 4 of the winners were female! That's about half, which is suspiciously even...

Let's look at points.

```python
points_url = 'https://prettygr.im/tdlm/taskmaster/normalized_scores.json?_labels=on'
points_response = requests.get(points_url)
points_json = points_response.json()
```

```python
points_json
```

```python
taskmaster_points = pd.DataFrame(columns=["id","task", "contestant", "base","adjustment","points","rank"])
```

```python
for row in points_json['rows']:
    tm_id = row['id']
    task = row['task']['label']
    contestant = row['contestant']
    contestant = contestant['label']
    base = row['base']
    adjustment = row['adjustment']
    rank = row['rank']
    points = row['points']
    
    new_row = {"id":tm_id,"task": task, "contestant": contestant, "base":base, "adjustment":adjustment, "points":points,"rank":rank}
    
    taskmaster_points = pd.concat([taskmaster_points, pd.DataFrame([new_row])], ignore_index=True)
```
